```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

x = np.array([1,2,2,3,4,4,5,6]).reshape(-1, 1)
y = np.array([2,3,4,4,4,6,5,7])

modelo = LinearRegression()
modelo.fit(x, y)

plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x, modelo.predict(x), color='red', label='Lnea de regresin')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Regresin Lineal Simple")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

value_x = [1,2,2,3,4,4,5,6]

value_y = [2,3,4,4,4,6,5,7]

data_dict = { "X":value_x,
              "Y":value_y}
df = pd.DataFrame(data_dict)
df["XY"] = df["X"] * df["Y"]
df["X**2"] =  df["X"]**2
df

total_x = df["X"].sum()
total_y = df["Y"].sum()
total_xy = df["XY"].sum()
total_x2 = df["X**2"].sum()

fila_total_dict = {"X": total_x, "Y": total_y, "XY": total_xy, "X**2": total_x2}

fila_total_df = pd.DataFrame([fila_total_dict])
df_con_total = pd.concat([df, fila_total_df], ignore_index=True)
df_con_total
```
