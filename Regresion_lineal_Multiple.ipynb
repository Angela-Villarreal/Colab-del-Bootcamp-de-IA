```python
import pandas as pd

value_y = [78.5,74.3,104.3,87.6,95.9,109.2,102.7,72.5,93.1,115.9,83.8,113.3,109.4]

value_x1 = [7,1,11,11,7,11,3,1,2,21,1,11,10]
value_x2 = [26,29,56,31,52,55,71,31,54,47,40,66,68]
value_x3 = [6,15,8,8,6,9,17,22,18,4,23,9,8]
value_x4 = [60,52,20,47,33,22,6,44,22,26,34,12,12]

data_dict = {"Y":value_y,
             "X1":value_x1,
             "X2":value_x2,
             "X3":value_x3,
             "X4":value_x4}
df = pd.DataFrame(data_dict)
df
```

```python
df["Y"]=1
df
```

```python
matriz_x = pd.DataFrame(df)
matriz_x
```

```python
df_transpuesto = df.T
df_transpuesto
```

```python
matriz_x_ = pd.DataFrame(df_transpuesto)
matriz_x_
```

```python
matriz_x_x = matriz_x_.dot(matriz_x)
matriz_x_x
```

```python
import numpy as np
matriz_x_x_inv = np.linalg.inv(matriz_x_x)
print(matriz_x_x_inv)
```

```python
import numpy as np
lista_y = np.array(value_y)
matriz_y = lista_y.reshape(-1, 1)
print(matriz_y)
```

```python
beta= matriz_x_x_inv.dot(matriz_x_) .dot(matriz_y)
print(beta)
```

```python
tabla_beta = pd.DataFrame(beta)
tabla_beta
```
