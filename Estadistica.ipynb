# **con 3 datos**

```python
import pandas as pd

value_nombre = ["Andres","Jose","Sofia"]
print(value_nombre)

value_peso = [58,65,72]
print(value_peso)

value_altura = [1.65,1.68,1.75]
print(value_altura)

data_dict = { "name":value_nombre,
              "weight":value_peso,
              "height":value_altura}
print(data_dict)

df = pd.DataFrame(data_dict)
print(df)
df
```

```python
df["bmi"] = df["weight"] / (df["height"]**2)
df
```

```python
df.describe()
```
