```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Datos: tamaÃ±o en mÂ² y su precio en dÃ³lares

x = np.array([30, 50, 60, 80, 100]).reshape(-1, 1)  # tamaÃ±o (mÂ²)
y = np.array([100000, 150000, 180000, 200000, 250000])  # precio

modelo = LinearRegression()
modelo.fit(x, y)

prediccion = modelo.predict(np.array([[70]]))  # predecimos precio para 70 mÂ²
print("Precio estimado para 70 mÂ²:", int(prediccion[0]), "USD")

plt.scatter(x, y, color='blue', label='Datos reales')  # puntos reales
plt.plot(x, modelo.predict(x), color='red', label='LÃ­nea de regresiÃ³n')  # lÃ­nea ajustada
plt.xlabel("TamaÃ±o (mÂ²)")
plt.ylabel("Precio (USD)")
plt.title("RegresiÃ³n Lineal Simple")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Simulamos 100 personas con experiencia entre 0 y 20 aÃ±os
np.random.seed(0)  # para resultados reproducibles
experiencia = np.random.uniform(0, 20, 100).reshape(-1, 1)  # aÃ±os de experiencia

# Ingreso mensual (base + experiencia * coeficiente + algo de ruido aleatorio)
ingreso = 20000 + experiencia * 3000 + np.random.normal(0, 5000, (100, 1))  # ingreso con ruido

modelo = LinearRegression()
modelo.fit(experiencia, ingreso)

ingreso_predicho = modelo.predict(experiencia)
plt.figure(figsize=(10, 6))
plt.scatter(experiencia, ingreso, color='skyblue', label='Datos reales')
plt.plot(experiencia, ingreso_predicho, color='red', linewidth=2, label='LÃ­nea de regresiÃ³n')
plt.xlabel("AÃ±os de experiencia")
plt.ylabel("Ingreso mensual (USD)")
plt.title("RegresiÃ³n lineal")
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Simulamos 100 personas con experiencia entre 0 y 20 aÃ±os
np.random.seed(0)  # para resultados reproducibles
experiencia = np.random.uniform(0, 20, 100).reshape(-1, 1)  # aÃ±os de experiencia

# Ingreso mensual (base + experiencia * coeficiente + algo de ruido aleatorio)
ingreso = 20000 + experiencia * 3000 + np.random.normal(0, 5000, (100, 1))  # ingreso con ruido

modelo = LinearRegression()
modelo.fit(experiencia, ingreso)

ingreso_predicho = modelo.predict(experiencia)
plt.figure(figsize=(10, 6))
plt.scatter(experiencia, ingreso, color='skyblue', label='Datos reales')
plt.plot(experiencia, ingreso_predicho, color='red', linewidth=2, label='LÃ­nea de regresiÃ³n')
plt.xlabel("AÃ±os de experiencia")
plt.ylabel("Ingreso mensual (USD)")
plt.title("RegresiÃ³n lineal")
plt.legend()
plt.grid(True)
plt.show()

# Â¿CuÃ¡nto ganarÃ­a una persona con 10 aÃ±os de experiencia?
print("Ingreso estimado para 10 aÃ±os de experiencia:", int(modelo.predict([[10]])[0][0]), "USD")
```
